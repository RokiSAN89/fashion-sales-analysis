# 📊 Fashion Sales Analysis – Notebook
This notebook explores global fashion retail sales data by category and country from 2023 to 2025.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
```

## 🔹 Load the dataset

```python
# Example loading (replace with actual path)
df = pd.read_csv('data/global_fashion_sales.csv')
df.head()
```

## 🔹 Sales by Category

```python
# Example plot
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x='Date', y='Masculine', label='Masculine')
sns.lineplot(data=df, x='Date', y='Feminine', label='Feminine')
sns.lineplot(data=df, x='Date', y='Children', label='Children')
plt.title('Global Fashion Sales by Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 🔹 Sales by Country (excluding China)

```python
plt.figure(figsize=(12,6))
for country in ['USA', 'Germany', 'France', 'Portugal']:
    sns.lineplot(data=df, x='Date', y=country, label=country)
plt.title('Fashion Sales by Country (Excluding China)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
